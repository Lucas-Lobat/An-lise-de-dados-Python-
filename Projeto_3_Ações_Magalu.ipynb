# Analise das ações da Magalu

```python
# Importando Libs Necessárias

# Libs para modelagens e matrizez
import numpy as np
import pandas as pd

# Libs para analises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avidos
warnings.filterwarnings('ignore')

```

```python
# Lendo os dados
Base_Dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')
```

```python
Base_Dados.head()
```

```python
# Verificando a quantidade de linhas e colunas
Base_Dados.shape
```

```python
Base_Dados.info()
```

```python
# Series temporais (Anexando a coluna "Data" no index)
Dados =Base_Dados.set_index('Data')

Dados.head()
```

```python
# Analisando as ações de fechamento da Magalu

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.title('Analise das Ações da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados['Fechamento']);

plt.xlabel('Período da cotação')
plt.ylabel('Valor da cotação (R$)');
```

```python
# Verificando as 5 últimas linhas de dados.
Dados.tail() 
```

```python
# Verificando a Media movel e Tendência em relação ao fechamento das ações

Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.title('Analise das Ações da Magalu - Fechamento', fontsize=15, loc='left')

plt.plot(Dados.index, Dados['Fechamento']);
plt.plot(Dados.index, Media_Movel)
plt.plot(Dados.index, Media_Tendencia)

plt.xlabel('Período da cotação')
plt.ylabel('Valor da cotação (R$)');
```

```python
Base_Dados['Mes'] = Base_Dados['Data'].dt.month

Base_Dados.head()
```

```python
# Analizando as variações mensais com o boxplot

plt.figure(figsize=(13,5))
sns.boxplot(data=Base_Dados, x='Mes', y='Fechamento')
```

```python
Base_Dados.groupby(by=['Mes']).describe()['Fechamento']
```

```python
# Gráfico coumente usado em financeiras
Grafico = go.Figure(
    data=[
        go.Candlestick(
        x= Dados.index,
        open= Dados['Abertura'],
        high= Dados['Maior'],
        low=Dados['Menor'],
        close= Dados['Fechamento']
    )]
)

Grafico.update_layout( xaxis_rangeslider_visible=False)
Grafico.show()
```
